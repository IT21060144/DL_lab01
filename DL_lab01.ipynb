2. Create and visualize a 100,000x1 array from the same exponential distribution

```python
import matplotlib.pyplot as plt

exp_array = np.random.exponential(scale=1.0, size=(100000, 1))

# Plotting the histogram for the exponential distribution
plt.hist(exp_array, density=True, bins=100, histtype="step", color="blue", label="Exponential")
# Adding uniform and normal distributions for comparison
plt.hist(np.random.uniform(low=-2.5, high=2.5, size=100000), density=True, bins=100, histtype="step", color="green", label="Uniform")
plt.hist(np.random.normal(loc=0, scale=1, size=100000), density=True, bins=100, histtype="step", color="red", label="Normal")
plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc="upper left")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

3. Plot Z=X2+Y2Z=X2+Y2 in a 3D plot

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = x**2 + y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, cmap='viridis')
ax.set_title('$Z = X^2 + Y^2$')
plt.show()
```

4. Calculate Pearson and Spearman correlations, and visualize with heatmaps

```python
import seaborn as sns
import pandas as pd

# Assuming seaborn tutorial data is loaded into a DataFrame `df`
# df = pd.read_csv('path_to_seaborn_tutorial_data.csv')
# Here's a dummy data for demonstration:
data = {
    'HP': np.random.rand(100),
    'Attack': np.random.rand(100),
    'Defense': np.random.rand(100),
    'Sp. Atk': np.random.rand(100),
    'Sp. Def': np.random.rand(100),
    'Speed': np.random.rand(100)
}
df = pd.DataFrame(data)

# Calculate Pearson and Spearman correlations
pearson_corr = df.corr(method='pearson')
spearman_corr = df.corr(method='spearman')

# Visualize with heatmaps
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

sns.heatmap(pearson_corr, annot=True, ax=ax[0], cmap='coolwarm')
ax[0].set_title('Pearson Correlation')

sns.heatmap(spearman_corr, annot=True, ax=ax[1], cmap='coolwarm')
ax[1].set_title('Spearman Correlation')

plt.show()
```

1. Load data and filter specific columns

```python
# Load the Au_nanoParticle_dataset.csv into a DataFrame
# df = pd.read_csv('path_to_Au_nanoParticle_dataset.csv')
# Here's a dummy data for demonstration:
data = {
    'N_total': np.random.rand(100),
    'N_bulk': np.random.rand(100),
    'N_surface': np.random.rand(100),
    'R_avg': np.random.rand(100),
    'Feature1': np.random.rand(100),
    'Feature2': np.random.rand(100)
}
df = pd.DataFrame(data)

# Filter columns
filtered_df = df[['Feature1', 'Feature2']]
```

2. Display the first 20 samples of the dataframe

3. Calculate mean, standard deviation, and quartiles

```python
mean_values = df.mean()
std_values = df.std()
quartiles = df.quantile([0.25, 0.5, 0.75])

print("Mean values:\n", mean_values)
print("Standard deviation values:\n", std_values)
print("Quartile values:\n", quartiles)
```

4. Plot histograms for each feature in a 1x4 layout

```python
fig, axs = plt.subplots(1, 4, figsize=(20, 5))

features = ['N_total', 'N_bulk', 'N_surface', 'R_avg']
for i, feature in enumerate(features):
    axs[i].hist(df[feature], bins=20, color='blue', alpha=0.7)
    axs[i].set_title(f'Histogram of {feature}')

plt.tight_layout()
plt.show()
```

5. Visualize scatter plots and histograms using pairplot

```python
sns.pairplot(df[['N_total', 'N_bulk', 'N_surface', 'R_avg']])
plt.show()
```

6. Customize PairGrid plots

```python
# Custom PairGrid
g = sns.PairGrid(df[['N_total', 'N_bulk', 'N_surface', 'R_avg']])
g.map_upper(sns.histplot)
g.map_diag(sns.histplot, kde=True)
g.map_lower(sns.kdeplot)
plt.show()
```
